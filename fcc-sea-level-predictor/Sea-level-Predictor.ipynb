```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
import numpy as np
```

```python
df=pd.read_csv('epa-sea-level.csv')
df
```

```python
fig, ax =plt.subplots(figsize=(8,8))
plt.scatter(data=df, x='Year', y='CSIRO Adjusted Sea Level',edgecolors='darkblue')
ax.set_xlabel('Year')
ax.set_ylabel('CSIRO Adjusted Sea Level')
```

```python
reg=linregress(df['Year'], df['CSIRO Adjusted Sea Level'])
```

```python
ax.set_aspect('auto')
x=np.arange(1880,2055)
ax.plot(x,x*reg.slope+reg.intercept, color='red', linewidth=2)
fig
```

```python
x=np.arange(2000, 2060)
f, axes=plt.subplots(figsize=(8,8))
axes.plot(x,x*reg.slope+reg.intercept, color='red')
axes.set_xlabel('Year')
axes.set_ylabel('Sea Level(Inches)')
axes.set_title('Rise in Sea Level')

plt.savefig('riseInsealevel.png')
```
